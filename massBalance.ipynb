I bilanci di massa consentono di verificare, in equivalenti d'acqua, se nella regione presa in considerazione alla fine della stagione d'ablazione si ha avuto un aumento oppure una diminuzione della massa disponibile. Il bilancio di massa è dato dalla somma algebrica di due termini, entrambi espressi in equivalenti d'acqua, i quali quantificano l'accumulo nevoso in water equivalent e l'ablazione.

$$
MB\,=\,S_{WE}\,+\,AM
$$

Una tecnica per misurare l'ablazione consiste nel porre delle paline ablatometriche, che consentono di valutare lo spessore nevoso e di ghiaccio che si va via va perdendo durante la stagione d'ablazione.
